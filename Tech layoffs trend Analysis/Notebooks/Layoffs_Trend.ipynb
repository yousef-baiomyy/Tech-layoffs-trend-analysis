```python
numeric_cols = ['layoffs_count', 'layoff_percentage', 'open_roles', 
                 'stock_growth_percent', 'revenue_growth_percent', 
                 'salary_budget_change']

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))
axes = axes.flatten()
for i, col in enumerate(numeric_cols):
    sns.boxplot(y=df[col], ax=axes[i], color='skyblue')
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
# we had many outliers in the first 3 figuers 
```

```python
top_layoffs = df[df['layoffs_count'] > 15000]
print(top_layoffs['company_size'].value_counts())
print(top_layoffs['industry'].value_counts())
print(top_layoffs['year'].value_counts())
```

```python
plt.figure(figsize=(10, 7))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Heatmap ')
plt.show()
```

```python
month_map = {'Jan':1,'Feb':2,'Mar':3,'Apr':4,'May':5,'Jun':6,
             'Jul':7,'Aug':8,'Sep':9,'Oct':10,'Nov':11,'Dec':12}
df['month_num'] = df['month'].map(month_map)
df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month_num'].astype(str) + '-01')
```

```python
monthly_trend = df.groupby('date')['layoffs_count'].sum().reset_index()
```

```python
plt.figure(figsize=(16,6))
plt.plot(monthly_trend['date'], monthly_trend['layoffs_count'], marker='o')
plt.title('total layoffs')
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.show()
```

```python
industry_layoffs = df.groupby('industry')['layoffs_count'].sum().sort_values(ascending=False)
print(industry_layoffs)

country_layoffs = df.groupby('country')['layoffs_count'].sum().sort_values(ascending=False)
print(country_layoffs)

reason_layoffs = df.groupby('reason_for_layoffs')['layoffs_count'].agg(['sum','mean','count']).sort_values('sum', ascending=False)
print(reason_layoffs)
```

```python

fig, axes = plt.subplots(1, 2, figsize=(16,6))
industry_layoffs.plot(kind='bar', ax=axes[0], color='steelblue', title='Layoffs by Industry')
country_layoffs.plot(kind='bar', ax=axes[1], color='coral', title='Layoffs by Country')
plt.tight_layout()
plt.show()
```
